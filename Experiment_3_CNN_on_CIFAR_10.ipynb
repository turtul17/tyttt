````python
# Train and evaluate a Convolutional Neural Network (CNN) for image classification using CIFAR-10 dataset.
````

````python
import torch
import torchvision
import torchvision.transforms as transforms           # for importing libraries
import matplotlib.pyplot as plt
````

````python
transform = transforms.ToTensor()                     # for data preprocessing
````

````python
trainset = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,                                       # for data loading
    download=True,
    transform=transform
)
````

````python
trainloader = torch.utils.data.DataLoader(
    trainset, batch_size=4, shuffle=True              # data loader
)                                                     # instead of loading all images i'll load only 4 imgs
````

````python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(6 * 14 * 14, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = x.view(-1, 6 * 14 * 14)
        x = self.fc1(x)
        return x
````

````python
def forward(self, x):
    x = self.pool(F.relu(self.conv1(x)))
    x = x.view(-1, 6*14*14)                           # forword pass
    x = self.fc1(x)
    return x
````

````python
criterion = nn.CrossEntropyLoss()
````

````python
model = CNN()
````

````python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)
````

````python
for epoch in range(2):   # keep 2 for fast run
    for inputs, labels in trainloader:

        optimizer.zero_grad()   # clear old gradients

        outputs = model(inputs) # forward pass
        loss = criterion(outputs, labels)  # calculate loss

        loss.backward()   # backpropagation
        optimizer.step()  # update weights

    print(f"Epoch {epoch+1} completed")
````

````python
print(model)
````

````python
correct = 0
total = 0

with torch.no_grad():
    for inputs, labels in trainloader:
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print("Accuracy:", 100 * correct / total)
````
